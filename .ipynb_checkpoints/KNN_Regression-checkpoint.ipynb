# Introduction

Simple KNN algorithm (myknn) for regression is implemented in this notebook. Simulated data was used to examine the implemented algorithm. Another set of simulated data was used to demonstrate the curse of dimensionality in KNN regression.

# KNN implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
%matplotlib inline
# helper function for calculating distant
# return indices of the knn
def _find_nnindices(_x, xtrain, k):
    '''
    Returns the row indices of k nearest neighbors
    
    Parameters:
    _x: data item of interest
    xtrain: training data
    k: number of neighbors
    
    Returns:
    np.array of shape (n, k), where n is the number of input data items
    '''
    return ((xtrain - _x)**2).sum(axis = 1).argsort()[:k]

# knn function
def myknn(x, xtrain, ytrain, k):
    '''
    Returns the estimated Y value of KNN regression
    
    Parameters:
    x: np.array data item of interest
    xtrain: np.array training data
    ytrain: np.array target variable 
    k: int number of neighbors
    
    Returns:
    np.array of shape (n), where n is the number of input data items
    '''
    nns = np.apply_along_axis(_find_nnindices, 1, x, xtrain=xtrain, k=k)
    prediction = ytrain[nns,].mean(axis=1)
    return prediction
```

To examine the algorithm, we'll use simulated data.

The covariate of training data has n=1000, with p=5 from independent standard Normal distribution.  
And Y is calculated by:  
$$Y=X_{1}+0.5*X_{2}−X_{3}+\epsilon$$
with i.i.d. standard normal error $\epsilon$. 
  
The algorithm is tested using the first 500 observations as the training data and the rest as testing data. 
We'll use k = 5 for prediction and evaluate the prediction accuracy using mean squared error
$$\frac{1}{N}\Sigma_{i}(y_{i}-\hat{y}_{i})^2$$


```python
# data generation
n = 1000
p = 5
np.random.seed(seed=2)
X = np.array(np.random.normal(0, 1, (n, p)))
Y = (X[:,0] + 0.5 * X[:,1] - X[:,2]).reshape(n, 1) + np.random.normal(0, 1, (n, 1))

xtrain = X[:500,]
xtest = X[500:,]
ytrain = Y[:500,]
ytest = Y[500:,]

# make prediction
pred = myknn(xtest, xtrain, ytrain, 5)

# calculate MSE
def calc_mse(ytest, pred):
    return sum((ytest - pred)**2)/pred.shape[0]

MSE = calc_mse(ytest, pred)
print('MSE for {} nearest neighbors model: {}'.format(5, round(MSE[0], 3)))
```

We'll further explore k from 1 to 10. The result is plotted using the degrees of freedom as the horizontal axis.  

Degrees of freedom for KNN model is $$\frac{N}{k}$$
Where N is the number of data items, and k the number of nearest neighbors.

```python
testing_mse = []

for k in range(1, 11, 1):
    pred = myknn(xtest, xtrain, ytrain, k)
    testing_mse.append(calc_mse(ytest, pred)[0])

df = []
for i in range(1, 11, 1):
    df.append(n/i)

plt.style.use('seaborn-whitegrid')
sns.set(font_scale=1.5)
fig = plt.figure(figsize=(14, 4))
ax = sns.lineplot(x=df, y=testing_mse, label='Testing MSE', marker='o')
x_lab = plt.xlabel('Degree of Freedom', weight='bold')
y_lab = plt.ylabel('Error Rate', weight='bold')
title = plt.title('DF v.s Errors', weight='bold')
legend = ax.legend()
```

From the plot, we see that with value of k ranges from 1 to 10, the associated degree of freedom spans from 100 to 1000. Our optimal tuning parameter k is 10, which achieves the lowest testing MSE 1.44.

# On the Curse of Dimensionality

Let’s consider a high-dimensional setting. With the model the same as above. We consider two cases that both generate an additional set of 95 covariates and append to the original covariate matrix:

- Generate another 95-dimensional covariate with all independent standard Gaussian entries  
- Generate another 95-dimensional covariate using the formula $X^TA$, where $X$ is the original 5-dimensional vector, and $A$ a 5×95 dimensional (fixed) matrix that remains the same for all observations. Where each data item in matrix $A$ are i.i.d. and follows a uniform $[0, 1]$ distibution.  

KNN was fit in both settings (with the total of 100 covariates) and the best K value was selected by trying a number of different K. 

```python
# data generation
n = 1000
p_2 = 95

# setting 1
np.random.seed(seed=2)
X_norm = np.array(np.random.normal(0, 1, (n, p_2)))
X_1 = np.c_[X, X_norm]
xtrain_1 = X_1[:500,]
xtest_1 = X_1[500:,]

testing_mse_1 = []
for k in range(1, 100, 1):
    pred = myknn(xtest_1, xtrain_1, ytrain, k)
    testing_mse_1.append(calc_mse(ytest, pred)[0])

# setting 2
np.random.seed(seed=2)
X_uniform = X @ np.array(np.random.uniform(0, 1, (5, p_2)))
X_2 = np.c_[X, X_uniform]
xtrain_2 = X_2[:500,]
xtest_2 = X_2[500:,]

testing_mse_2 = []
for k in range(1, 100, 1):
    pred = myknn(xtest_2, xtrain_2, ytrain, k)
    testing_mse_2.append(calc_mse(ytest, pred)[0])
```

```python
print('Best k for setting 1: {}, MSE: {}'.format((testing_mse_1.index(min(testing_mse_1)) + 1), round(min(testing_mse_1), 3)))
print('Best k for setting 2: {}, MSE: {}'.format((testing_mse_2.index(min(testing_mse_2)) + 1), round(min(testing_mse_2), 3)))
```

```python
plt.style.use('seaborn-whitegrid')
sns.set(font_scale=1.5)
fig = plt.figure(figsize=(14, 4))
ax = sns.lineplot(x=range(1, 100, 1), y=testing_mse_1, label='Setting 1', marker='o')
ax_2 = sns.lineplot(x=range(1, 100, 1), y=testing_mse_2, label='Setting 2', marker='o')
x_lab = plt.xlabel('K', weight='bold')
y_lab = plt.ylabel('Error Rate', weight='bold')
title = plt.title('K v.s MSE', weight='bold')
legend = ax.legend()
```

Best k for setting 1 is 19, which achieves testing MSE of 2.938. On the other hand, best k for setting 2 is 6 which achieves testing MSE of 1.71.  

Setting 2 performs better in general. In setting 1, we added a bunch of not useful covariates which are random noise, therefore, we did not increase the predictive power of the model by introducing these new covariates, in fact, we are lowering the chance that we found 'good representatives' from the training data in the k nearest neighbors.  

Even though in setting 2, we added colinearity between covariates, which is often a bad practice in ML problems, we do not suffer form curse of dimensionality that much comaparing to setting 1, because the new set of covariates are actually linear combinations of the original covariates, hence the rank of the covariate matrix does not increase.
